time periods

```python
year_month = dataframe[[('سال', ''),('ماه', '')]]
year_monthls = []
for i , row in year_month.iterrows():
    if tuple([row[('سال', '')] , row[('ماه', '')]]) not in year_monthls:
        year_monthls.append(tuple([row[('سال', '')] , row[('ماه', '')]]))

with Session.begin()as con:
    for y_m in year_monthls:
        new_time_priod = time_priod(y_m[0],y_m[1])
        con.add(new_time_priod)
    con.commit()
```

test_parameter

```python
with Session.begin() as con:
    testsquery = con.query(tests).all()
    parametersquery = con.query(parameters).all()
    for col in dataframe.columns[4:]:
        ntest_id = None
        npar_id = None
        for test in testsquery:
            if test.test_name == col[0][-5:]:
                ntest_id = test.test_id
                break
        if ntest_id:
            for par in parametersquery:
                if col[1] == par.parameter_name:
                    npar_id = par.parameter_id
                    break
            if npar_id:
                ntest_par = test_parameter(ntest_id , npar_id)
                con.add(ntest_par)
            else:
                print("par not exists")
        else:
            print("test not exist")
    con.commit()
        
```

work orders

```python
with Session.begin() as con:
    testsquery = con.query(tests).all()
    citiesquery = con.query(cities).all()
    time_periodsquery = con.query(time_priod).all()
    
    for vtest in testsls:
        nctid = None
        ntimeid = None
        ntstid = None
        for i , row in dataframe[dataframe.columns[:4]].iterrows():
            for ct in citiesquery:
                if ct.name == row[('مدیریت برق شهرستان', '')]:
                    nctid = ct.city_id
                    break
            else:
                print("ct not found")
            if nctid:
                for time in time_periodsquery:
                    if time.month == row[('ماه', '')] and time.year == row[('سال', '')]:
                        ntimeid = time.period_id
                        break
                else:
                    print("time not found")
                for ntest in testsquery:
                    if vtest == ntest.test_name:
                        ntstid = ntest.test_id

            if nctid and ntimeid and ntstid:
                nworkorders = work_orders(ntstid , nctid , ntimeid)
                con.add(nworkorders)
    con.commit()           

```

work_order_stats


```python
df = dataframe.set_index([('مدیریت برق شهرستان', ''),
 ('کد امور', ''),
 ('سال', ''),
 ('ماه', '')])
```

```python
df.index.names = ['مدیریت برق شهرستان', 'کد امور', 'سال' , 'ماه']
```

```python
with Session.begin() as con:
    wordersquery = con.query(work_orders).all()
    tquery = con.query(tests).all()
    ctquery = con.query(cities).all()
    tpquer = con.query(time_priod).all()
    pquery = con.query(parameters).all()
    
    for row in df.index :
        ct_id = None
        tpr_id = None
        tst_id = None
        p_id = None
        wo_id = None


        ct = row[0]
        year = row[2]
        month = row[3]
        for ect in ctquery:
            if ect.name == ct:
                ct_id = ect.city_id
                break
        if ct_id:
            for index in df.loc[row].index:
                for tim in tpquer:
                    if tim.year == year and tim.month == month:
                        tpr_id = tim.period_id
                        break
                if tpr_id:
                    for tst in tquery:
                        if tst.test_name == index[0][-5:]:
                            tst_id = tst.test_id
                            break
                    for pr in pquery:
                        if pr.parameter_name == index[1]:
                            p_id = pr.parameter_id
                            break
                    if tst_id and p_id:
                        for wo in wordersquery:
                            if wo.test_id == tst_id and wo.city_id == ct_id and wo.period_id == tpr_id:
                                wo_id = wo.work_order_id
                                break
                        if wo_id:
                            nr = work_order_stats(work_order_id=wo_id,parameter_id=p_id,count=int(df[index].loc[row]))
                            print(type(nr.count))
                            con.add(nr)
                           





```
